# Muffin Vs Cupcake Prediction Using Recipe Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

## Reading our dataset using pandas

```python
recipe = pd.read_csv("recipes_muffins_cupcakes.csv")
recipe
```

## Changing our categorical Values "Muffin and Cupcake" into binary values

```python
label = np.where(recipe["Type"] == "Muffin", 0, 1)
```

```python
label
```

## Assigning dependent varibales to y and Independet variables to X

```python
X = recipe[["Flour", "Milk", "Sugar", "Butter", "Egg", "Baking Powder", "Vanilla", "Salt"]]
y = label
```

```python
X
```

```python
y
```

## Splitting training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Fitting our Logistic Regression Algorithm inside our training dataset

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
regressor.fit(X_train, y_train)
regressor
```
